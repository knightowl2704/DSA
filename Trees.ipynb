# Build Tree from Inorder and Postorder

```python
In = [4, 8, 2, 5, 1, 6, 3, 7] 
post = [8, 4, 5, 2, 6, 7, 3, 1]  
n = len(In) 

def buildtree(In,post,n):
    pIndex = [n-1]
    return buildtreeUtils(In,post,0,n-1,pIndex)

def buildtreeUtils(In,post,Instrt,Inend,pIndex):
    if Instrt > Inend:
        return None
    
    node = Node(post[pIndex[0]])
    pIndex[0] = pIndex[0]-1
    
    if Instrt == Inend:
        return node
    
    iIndex = search(In,Instrt,Inend,node.value)
    
    node.right = buildtreeUtils(In,post,iIndex+1,Inend,pIndex)
    node.left = buildtreeUtils(In,post,Instrt,iIndex-1,pIndex)
    
    return node

def search(arr,start,end,value):
    for i in range(start,end):
        if arr[i] == value:
            break
    return i
        
```

# Lowest Common Ancestor

```python
def LCA(root,n1,n2):
    if root is None:
        return None
    
    if root.value == n1 or root.value == n2:
        return root
    
    left_LCA = LCA(root.left,n1,n2)
    right_LCA = LCA(root.right,n1,n2)
    
    if left_LCA or right_LCA:
        return root
    
    return left_LCA if left_LCA is not None else right_LCA


```

# Count Leaf Nodes in tree

```python
def getleafnodes(root):
    if root is None:
        return 0
    if root.left is None and root.right is None:
        return 1
    return getleafnodes(root.right)+getleafnodes(root.left)
```

# Maximum Depth of the Tree

```python
def max_depth(root):
    if root is None:
        return 0
    else:

        ldepth = max_depth(root.left)
        rdepth = max_depth(root.right)
        
        if ldepth > rdepth:
            return ldepth+1
        else:
            return rdepth+1
```

# Check if tree is Symmetric (Important)

```python
def isSymmetric(root):
    if root is None:
        return True
    return isSymmetrictwo(root.left,root.right)
def isSymmetrictwo(left,right):
    if left is None or right is None:
        return left == right
    
    if left.value != right.value:
        return False
    
    return isSymmetrictwo(left.right,right.left) and isSymmetrictwo(left.left, right.right)
```

# Level Order Traversal

```python
import collections
def levelOrder(root):
    if root is None:
        return []
    
    q = [(root,0)]
    h = collections.defaultdict(list)
    
    while q:
        node, lvl = q.pop(0)
        if node:
            h[lvl].append(node.value)
            q.append((node.left,lvl+1))
            q.append((node.right,lvl+1))
            print(h)
    return h.values()
```

# Minimum Depth of Binary Tree (Important) using Deque

```python
from collections import deque
def minDepth(root):
    if root is None:
        return 0
    
    q = deque(((root,1),))
    while q:
        node,lvl = q.popleft()
        if node:
            if node.left is None and node.right is None:
                return lvl
            q.append((node.left,lvl+1))
            q.append((node.right,lvl+1))
                    
    
    
```

# Maximum Depth of Binary Tree (Important) using Deque

```python
# Algorithm:
#     Create ((root,level))
#     maxlevelnow initialise to root level
#     while q:
#         pop left and check if node 
#         if yes: 
#             if node is leaf node?
#                 update maxlevelnow


def maxDepth(root):
    if root is None:
        return 0
    q = deque(((root,1),))
    maxnow = 1
    while q:
        node,lvl = q.popleft()
        if node:
            if node.left is None and node.right is None:
                if lvl > maxnow:
                    maxnow = lvl
            q.append((node.left,lvl+1))
            q.append((node.right,lvl+1))
    return maxnow
            
```

# Balanced Tree (Important)

```python
#                            1  <--------------- root2
#                          /   \
#                         2     3
#                       /   \
#                      4     5

# Algorithm:
#     DFS over the left and right sub tree return the values 
#     check if abs(left-right) <=1

def isBalanced(root):
    if root is None:
        return True
    
    left = dfs(root.left)
    right = dfs(root.right)
    p = abs(left-right)
    
    if p <=1:
        return isBalanced(root.left) and isBalanced(root.right)
    else:
        return False
    
def dfs(root):
    if root is None:
        return 0
    
    return max(dfs(root.left),dfs(root.right))+1
```

# Path Sum

```python
def hasPathSum(root,sumreq):
    if root is None:
        return False
    if root.value is sumreq and (root.left is None and root.right is None):
        return True
    return hasPathSum(root.right, sumreq-root.value) or hasPathSum(root.left,sumreq-root.value)
```
