```python
class Node:
    def __init__(self,value):
        self.value = value
        self.left = None
        self.right = None
                  
        
```

```python
root = Node(1) 
root.left      = Node(2) 
root.right     = Node(3) 
root.left.left  = Node(4) 
root.left.right  = Node(5) 
```

```python
def inorder(root):
    if root:
        inorder(root.left)
        print(root.value)
        inorder(root.right)
```

```python
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.value)
        
```

# Lowest Common Ancestor

```python
def LCA(root,n1,n2):
    if root is None:
        return None
    
    if root.value == n1 or root.value == n2:
        return root
    
    left_LCA = LCA(root.left,n1,n2)
    right_LCA = LCA(root.right,n1,n2)
    
    if left_LCA or right_LCA:
        return root
    
    return left_LCA if left_LCA is not None else right_LCA


```
